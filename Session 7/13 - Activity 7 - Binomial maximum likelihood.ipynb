# Activity 7: Binomial maximum likelihood #

We will now experiment with calculating maximum likelihood estimates (MLE) for the probability of default/Bad using a binomial function, i.e. only looking at Y-variable distribution, ignoring any X-variables. Later on we will learn that MLE models can only be compared if they are estimated on the same dataset, and this activity illustrates how each sample may produce different estimates. 

For the whole dataset P(Bad) = 0.3. You will take repeatedly small samples (n=10), and for each sample calculate a MLE estimate for P(Bad). Since we have already established that a sample proportion is a MLE estimate for a population proportion (or P(Bad) in our case), we will simply calculate it by taking a mean of the target variable.

```python
import statistics
import matplotlib.pyplot as plt
import pandas as pd

credit_data = pd.read_csv('credit_regress.csv')
# We only need a target variable Bad_2 for this exercise

status=credit_data[["Bad_2"]]
status.head()
```

```python
results = []
runs = []

# This will repeat the experiment for 100 times
for x in range(100):   
    run = status['Bad_2'].sample(n=10)
    p = statistics.mean(run)
    runs.append(x)
    results.append(p)

plt.plot(runs, results)
plt.show()
```

The chart shows the MLE estimates (Y-axis) from 100 runs (X-axis). Note that the estimate depends on a specific sample (re-running this gives different results). Note also the spread of the estimates (min and max values). 

Repeat the same exercise for a larger sample size, n = 100. Note any change in the spread of the estimates.
